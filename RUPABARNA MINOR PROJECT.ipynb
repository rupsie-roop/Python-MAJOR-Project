## Grading System

```python
df['total marks']=df['math score']+df['reading score']+df['writing score']
```

```python
df['percentage']=df['total marks']/300*100
```

GRADING SCALE

90-100 : Grade O

80-89 : Grade E

70-79 : Grade A

60-69 : Grade B

50-59 : Grade C

40-49 : Grade D

0-39 : Grade F

```python
def grades(scores):
    if scores >= 90 and scores <= 100:
        return 'Grade O'
    elif scores >= 80 and scores <= 90:
        return 'Grade E'
    elif scores >= 70 and scores <= 80:
        return 'Grade A'
    elif scores >= 60 and scores <= 70:
        return 'Grade B'
    elif scores >= 50 and scores <= 60:
        return 'Grade C'
    elif scores >= 40 and scores <= 50:
        return 'Grade D'
    elif scores >= 0 and scores <= 40:
        return 'Grade F'
    
df['grades']=df['percentage'].apply( grades)
df.info()
```

```python
plt.title('GRADES',fontsize=20)
df['grades'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

In the pie-chart above, we find, maximum number of students have achieved Grade A followed by Grade B, Grade C, Grade E, Grade D Grade O and then Grade F.

```python
plt.figure(figsize=(20,10))
plt.title('GRADES AND GENDER')
sns.countplot(hue="gender",x="grades",data=df)
plt.show()
```

The above graph is on grades by different genders.

Number of females securing Grade O is considerably more compared to males.

Number of females securing Grade E is more compared to males.

Number of females securing Grade A is more compared to males.

Number of males securing Grade C is slightly more compared to females.
